```python
import random
import time
from Bio.Align.Applications import ClustalOmegaCommandline, MuscleCommandline
from Bio import AlignIO
import pandas as pd
import matplotlib.pyplot as plt

# Function to introduce mutations
def introduce_mutations(seq, k):
    nucleotides = ['A', 'T', 'C', 'G']
    seq_list = list(seq)
    for _ in range(k):
        pos = random.randint(0, len(seq_list) - 1)
        new_nucleotide = random.choice([n for n in nucleotides if n != seq_list[pos]])
        seq_list[pos] = new_nucleotide
    return ''.join(seq_list)

# Original sequence
X = "AGCT" * 100  # A simple repetitive sequence for illustration

# Generate 10 variations with k=10 mutations each
variations = [introduce_mutations(X, 10) for _ in range(10)]

# Save the sequences to a FASTA file
with open("sequences.fasta", "w") as output_handle:
    for i, seq in enumerate(variations):
        output_handle.write(f">X{i+1}\n{seq}\n")

# Measure Clustal Omega time
start_time = time.time()
clustalomega_cline = ClustalOmegaCommandline(infile="sequences.fasta", outfile="clustal_output.aln", verbose=True, auto=True, force=True, outfmt="fa")
stdout, stderr = clustalomega_cline()
clustal_time = time.time() - start_time

# Measure MUSCLE time
start_time = time.time()
muscle_cline = MuscleCommandline(input="sequences.fasta", out="muscle_output.aln")
stdout, stderr = muscle_cline()
muscle_time = time.time() - start_time

# Read alignments
clustal_align = AlignIO.read("clustal_output.aln", "fasta")
muscle_align = AlignIO.read("muscle_output.aln", "fasta")

# Compare the alignments
print("Clustal Omega alignment length:", clustal_align.get_alignment_length())
print("MUSCLE alignment length:", muscle_align.get_alignment_length())

# Present the results
data = {
    "Algorithm": ["Clustal Omega", "MUSCLE"],
    "Time (seconds)": [clustal_time, muscle_time],
    "Alignment Length": [clustal_align.get_alignment_length(), muscle_align.get_alignment_length()]
}

df = pd.DataFrame(data)
print(df)

# Plot the results
df.plot(x="Algorithm", y="Time (seconds)", kind="bar", legend=False)
plt.ylabel("Time (seconds)")
plt.title("Time Taken by Algorithms")
plt.show()

df.plot(x="Algorithm", y="Alignment Length", kind="bar", legend=False)
plt.ylabel("Alignment Length")
plt.title("Alignment Length by Algorithms")
plt.show()
```

```python
import random
import time
from Bio.Align.Applications import ClustalOmegaCommandline, MuscleCommandline
from Bio import AlignIO, pairwise2
import pandas as pd
import matplotlib.pyplot as plt

# Function to introduce mutations
def introduce_mutations(seq, k):
    nucleotides = ['A', 'T', 'C', 'G']
    seq_list = list(seq)
    for _ in range(k):
        pos = random.randint(0, len(seq_list) - 1)
        new_nucleotide = random.choice([n for n in nucleotides if n != seq_list[pos]])
        seq_list[pos] = new_nucleotide
    return ''.join(seq_list)

# FRA10B_datasets sequence
original_seq = """ATATATATATATTTTTTTATATACATTATATATTTATATATAAATAAAATTTTTATTACATATATACATT
ATATACTTATACATAATTTTTTATTTTATTATATATTTATATATCTCTATTATATACATTATATAATTAT
ATATCATATATTTATATATGATATAATTATATATCATATATTTATATATGATATAATTATATATCATATA
TTTATATATGATATAATTATATATCATATATTTATATATGATATAATATATATGTATATAATATATAAAT
ATATATAAATATATAATATATATTTATATATATTTTATATATATACACATATTATATATGTGTATATTAT
ATAATATATACACATATTATATATGTGTATATTATATAATATACACATATATCATATAATTATATATTAT
ATAATATATATATTATATATGTATATATTATATATTATATAATGTATATATTATATATTATATAATGTAT
ATATTATATCATATATAATATATTATATATATGTATATATTATATATGATATATAATATATAATATATAT
GTATATATTATATATGATATAATATATACATATTATATATGTATATATTATATATGATATAATATATACA
TATTATATATGTATATATTATATATGATATAATATATACATATTATATGTATATATTATATAATATATAC
ATATATATGTATATATTATATATGTATATATTTATATACATATATAATATATGTATATATTATGTATATA
TTTATATACATATATTATATATAATATATATAATATATACGTATATAATATATAATACACTATATAATAT
ATACGTATATAATATATATATTAGATATAATATATACGTATATAATATATAATACACTATATAATATATA
CGTATATAATACATATATACTATATAATATATACGTATATAATACATATATACCATATAATATATACGTA
TATAATACATGTATTATATACGTATATAATATATATGTATACAATATATACGTATATAATATATATGTAT
ATATTATATAAATTATATATTATATATATGTATTATATAAATTATATATTTTTT"""

# Generate 10 variations with k=10 mutations each
variations = [introduce_mutations(original_seq, 10) for _ in range(10)]

# Save the sequences to a FASTA file
with open("sequences.fasta", "w") as output_handle:
    for i, seq in enumerate(variations):
        output_handle.write(f">X{i+1}\n{seq}\n")

# Measure Clustal Omega time
start_time = time.time()
clustalomega_cline = ClustalOmegaCommandline(infile="sequences.fasta", outfile="clustal_output.aln", verbose=True, auto=True, force=True, outfmt="fa")
stdout, stderr = clustalomega_cline()
clustal_time = time.time() - start_time

# Measure MUSCLE time
start_time = time.time()
muscle_cline = MuscleCommandline(input="sequences.fasta", out="muscle_output.aln")
stdout, stderr = muscle_cline()
muscle_time = time.time() - start_time

# Read alignments
clustal_align = AlignIO.read("clustal_output.aln", "fasta")
muscle_align = AlignIO.read("muscle_output.aln", "fasta")

# Custom substitution matrix
substitution_matrix = {
    ('A', 'A'): 10, ('C', 'C'): 10, ('G', 'G'): 10, ('T', 'T'): 10,
    ('A', 'C'): -1, ('A', 'G'): -1, ('A', 'T'): -1,
    ('C', 'A'): -1, ('C', 'G'): -1, ('C', 'T'): -1,
    ('G', 'A'): -1, ('G', 'C'): -1, ('G', 'T'): -1,
    ('T', 'A'): -1, ('T', 'C'): -1, ('T', 'G'): -1
}

# Calculate sum-of-pairs score
def sum_of_pairs_score(alignment, matrix):
    score = 0
    for i in range(len(alignment)):
        for j in range(i + 1, len(alignment)):
            seq1 = alignment[i].seq
            seq2 = alignment[j].seq
            for k in range(len(seq1)):
                pair = (seq1[k], seq2[k])
                if pair in matrix:
                    score += matrix[pair]
                else:
                    score += -2  # Penalty for gap or mismatch not in matrix
    return score

clustal_score = sum_of_pairs_score(clustal_align, substitution_matrix)
muscle_score = sum_of_pairs_score(muscle_align, substitution_matrix)

print("Clustal Omega sum-of-pairs score:", clustal_score)
print("MUSCLE sum-of-pairs score:", muscle_score)

# Present the results
data = {
    "Algorithm": ["Clustal Omega", "MUSCLE"],
    "Time (seconds)": [clustal_time, muscle_time],
    "Alignment Length": [clustal_align.get_alignment_length(), muscle_align.get_alignment_length()],
    "Sum-of-Pairs Score": [clustal_score, muscle_score]
}

df = pd.DataFrame(data)
print(df)

# Plot the results
df.plot(x="Algorithm", y="Time (seconds)", kind="bar", legend=False)
plt.ylabel("Time (seconds)")
plt.title("Time Taken by Algorithms")
plt.show()

df.plot(x="Algorithm", y="Alignment Length", kind="bar", legend=False)
plt.ylabel("Alignment Length")
plt.title("Alignment Length by Algorithms")
plt.show()

df.plot(x="Algorithm", y="Sum-of-Pairs Score", kind="bar", legend=False)
plt.ylabel("Sum-of-Pairs Score")
plt.title("Sum-of-Pairs Score by Algorithms")
plt.show()
```
